```python
metrics= pd.read_csv("../Public_Dataset/All_metrics_merger_heirarchial.csv")
metrics
```

```python
metrics =metrics[["assay", "method", "AUC"]]
metrics
```

```python
metrics_matrix= metrics.pivot(index=["assay"], columns="method")
metrics_matrix.columns = metrics_matrix.columns.droplevel(0)
metrics_matrix = metrics_matrix.reset_index()
metrics_matrix = metrics_matrix.reset_index(drop="True")
metrics_matrix = metrics_matrix.iloc[:-2, :]
metrics_matrix["assay"] = metrics_matrix["assay"].apply(int)
metrics_matrix = metrics_matrix[metrics_matrix.assay.isin(assay_GO_matrix.assay.unique())].reset_index(drop=True)
metrics_matrix.to_csv("metrics_matrix.csv", index=False)
metrics_matrix
```

```python
assay_GO_metrics_matrix = pd.merge(assay_GO_matrix, metrics_matrix)
assay_GO_metrics_matrix
```

```python
assay_GO_metrics_matrix["CP>MFP"] = assay_GO_metrics_matrix["CP"] > assay_GO_metrics_matrix["MFP"]

assay_GO_metrics_matrix["HM>MFP"] = assay_GO_metrics_matrix["Heirarchial Model"] > assay_GO_metrics_matrix["MFP"]
assay_GO_metrics_matrix["HM>CP"] = assay_GO_metrics_matrix["Heirarchial Model"] > assay_GO_metrics_matrix["CP"]

assay_GO_metrics_matrix["DM>MFP"] = assay_GO_metrics_matrix["Distance Based Merger"] > assay_GO_metrics_matrix["MFP"]
assay_GO_metrics_matrix["DM>CP"] = assay_GO_metrics_matrix["Distance Based Merger"] > assay_GO_metrics_matrix["CP"]
assay_GO_metrics_matrix["DM>HM"] = assay_GO_metrics_matrix["Distance Based Merger"] > assay_GO_metrics_matrix["Heirarchial Model"]
assay_GO_metrics_matrix
```
